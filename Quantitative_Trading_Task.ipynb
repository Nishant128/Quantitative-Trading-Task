```python
# importing all the important python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

- To maximize long-term returns, we need to find the optimal percentage of the stock price Trader G should risk on each trade. Here's the breakdown:

**Expected Value:**

Calculate the expected value of each trade:(Win% * Win Amount) + (Loss% * Loss Amount).
In this case, Expected Value = (40% * 2 rupees) + (60% * -1 rupee) = 0.2 rupees.


 **Risk Percentage:**

We need to relate the expected value to the risk taken (percentage of the stock price).
Let 'R' be the risk percentage as a decimal (e.g., 1% risk becomes 0.01).
The total capital risked per trade would be R * Stock Price.
Since the potential loss per trade is 1 rupee, we can set up an equation: Loss Amount = R * Stock Price = 1 rupee.


**Solving for Optimal Risk:**

Substitute the expected value into the previous equation: 0.2 rupees = R * Stock Price.
Solving for R, we get R = 0.2 rupees / Stock Price.
Unfortunately, we cannot determine the precise optimal risk without knowing the Stock Price.

Therefore, the optimal risk percentage for Trader G depends on the Stock Price. As the Stock Price rises, the optimal risk percentage (R) decreases proportionally. This makes sense: risking a larger percentage of a higher price stock could lead to catastrophic losses.

**What I Learned:**

- Risk-reward ratio alone isn't enough to define optimal risk. Understanding win/loss rates and expected value is crucial.
- The optimal risk percentage is dynamic and depends on market conditions (represented by Stock Price).
- This analysis highlights the importance of adjusting risk based on market context to maximize sustainable returns.

### Task 1
- Summary table along with suitable graph showing returns at different levels of percentage risk (consider risk of 1%, 2%, 5%, 10%, 15%, 20%, 40%, 80%) in a stock for the next 10 trades with 40% win rate.

```python
# Define variables
stock_price = 100  # Assuming a stock price of 100
win_rate = 0.4  # 40% win rate
loss_rate= 0.6
reward_risk_ratio = 2  # 2:1 reward to risk ratio
trade_count = 10  # 10 trades

# Calculate risk amounts for different percentages
risk_percentages = [0.01, 0.02, 0.05, 0.1, 0.15, 0.2, 0.4, 0.8]
risk_amounts = [risk_percentage * stock_price for risk_percentage in risk_percentages]

# Calculate expected returns for each risk level
expected_returns = []
for risk_amount in risk_amounts:
    win_amount = reward_risk_ratio * risk_amount
    loss_amount = -risk_amount
    expected_return = (win_rate * win_amount) + ((1 - win_rate) * loss_amount)
    expected_returns.append(expected_return)

# Create summary table
print("Summary Table:")
print("Risk % | Expected Return")
print("-------|---------------")
for i, (risk, return_) in enumerate(zip(risk_percentages, expected_returns)):
    print(f"{risk * 100:.2f}% | {return_:.2f}")

# Create graph
plt.figure(figsize=(10, 6))
plt.plot(risk_percentages, expected_returns, marker='o', linestyle='--', label='Expected Return')
plt.xlabel('Risk Percentage (%)')
plt.ylabel('Expected Return (Rupees)')
plt.title('Expected Return vs. Risk Percentage for 10 Trades (40% Win Rate)')
plt.grid(True)
plt.legend()
plt.show()
```
